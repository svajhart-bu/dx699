# 4. Storytelling With Data graph

Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "date": ["2026-01", "2026-01", "2026-01"],
    "competitor": ["A", "B", "C"],
    "revenue": [350000, 250000, 200000],
    "total_market": [1000000, 1000000, 1000000]
})

df["market_share"] = df["revenue"] / df["total_market"]

latest_date = df["date"].max()

pie_df = df[df["date"] == latest_date]

pie_df = (
    df.groupby("competitor", as_index=False)["market_share"]
    .sum()
)

pie_df["market_share_pct"] = pie_df["market_share"] * 100

plt.figure(figsize=(8, 8))

plt.pie(
    pie_df["market_share_pct"],
    labels=pie_df["competitor"],
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Market Share by Competitor")
plt.axis("equal")  # keeps pie circular

plt.show()
```
